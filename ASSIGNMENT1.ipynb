# PART 1

```python
import pandas as pd
import matplotlib as plt
from matplotlib import pyplot
import numpy as np
df = pd.read_csv('jsrt_metadata.csv')
print("Number of Patient of Each Disease=-\n")
print(df['diagnosis'].value_counts())
df['diagnosis'].value_counts().max()
mx = df['diagnosis'].value_counts().idxmax()
df['diagnosis'].replace(np.nan,mx,inplace =True)
plt.pyplot.title('Total Number of Patients of each Disease')
plt.pyplot.ylabel('No of Patients')
plt.pyplot.xlabel('Diseases')
plt.pyplot.hist(df['diagnosis'], color = 'yellow',ec='black')
```

# PART 2


```python
df.replace("non-nodule",np.nan,inplace = True)
df.replace("benign",np.nan,inplace = True)
df.dropna(subset = ["state"], inplace=True) 
plt.pyplot.title('Total Number of Patients of each Disease (Malignant)')
plt.pyplot.pie(df['diagnosis'].value_counts(),autopct = '%1.1f%%' , labels = ['lung cancer','lung cancer(adenocarcinoma)','lung cancer(bronchioloalveolar cell carcinoma)','metastasis of colon cancer','lung cancer(squamous cell carcinoma)','lung cancer(small cell carcinoma) ','lung canceror metastasis','lung cancer(adenosquamous carcinoma)','lung cancer(large cell carcinoma)','metastatic lung cancer','metastasis of ovarian cancer','etastasis of uterine cancer','susp. lung metastasis'] )

```
